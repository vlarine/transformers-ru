```python
from bertviz.pytorch_transformers_attn import BertModel, BertTokenizer, BertConfig
from bertviz.head_view import show
```

```python
%%javascript
require.config({
  paths: {
      d3: '//cdnjs.cloudflare.com/ajax/libs/d3/3.4.8/d3.min',
      jquery: '//ajax.googleapis.com/ajax/libs/jquery/2.0.0/jquery.min',
  }
});
```

```python
model_type = 'bert'
vocab_file = './rubert_cased_L-12_H-768_A-12_v1/vocab.txt'
model_file = './rubert_cased_L-12_H-768_A-12_v1/bert_model.bin'
config_file = './rubert_cased_L-12_H-768_A-12_v1/bert_config.json'
do_lower_case = False
config = BertConfig.from_json_file(config_file)
model = BertModel.from_pretrained(model_file, config=config)
tokenizer = BertTokenizer.from_pretrained(vocab_file, do_lower_case=do_lower_case)
```

```python
sentence_a = '   ,   .'
sentence_b = '  ,    ?'
show(model, model_type, tokenizer, sentence_a, sentence_b)
```
